```python
import requests
import pandas as pd

# 1. Cấu hình Endpoint cho dữ liệu Lịch sử (Archive API)
url = "https://archive-api.open-meteo.com/v1/archive"

# 2. Thiết lập tham số (Lấy khoảng 630 ngày để đảm bảo > 15.000 dòng)
# Toạ độ từ ảnh: 16.0678, 108.2208 (Đà Nẵng)
params = {
    "latitude": 16.0678,
    "longitude": 108.2208,
    "start_date": "2024-01-01",  # Ngày bắt đầu (Ví dụ)
    "end_date": "2026-01-01",    # Ngày kết thúc (~630 ngày sau)
    "hourly": ",".join([
        "temperature_2m",
        "relative_humidity_2m",
        "dew_point_2m",
        "surface_pressure",
        "rain",
        "showers",
        "snow_depth",
        "weather_code",
        "cloud_cover",
        "wind_speed_10m"
    ]),
    "timezone": "auto"
}

# 3. Gọi API
response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    
    # 4. Xử lý dữ liệu thành bảng (DataFrame)
    hourly_data = data['hourly']
    df = pd.DataFrame(hourly_data)
    
    # 5. Kiểm tra số lượng bản ghi
    print(f"Đã lấy thành công: {len(df)} bản ghi.")
    
    # 6. Lưu ra file CSV
    df.to_csv("weather_data_danang.csv", index=False)
    print("Đã lưu file: weather_data_danang.csv")
else:
    print("Lỗi:", response.status_code, response.text)
```
